```python
import matplotlib.pyplot as plt
import itertools
import statistics
```

```python
def insertionSort(array):

    comparations = 0
    swaps = 0
    instructions = 0
    for i in range(1,len(array)):
        temp = array[i]
        j = i - 1
        while j > -1 and temp < array[j]:
            comparations += 1
            array[j+1] = array[j]
            j = j-1
            swaps += 1
            instructions += 3
        instructions += 1
        array[j+1] = temp
        instructions += 4

        if j != i:
            comparations += 1

    instructions += 1
    result = []
    result.append(instructions)
    result.append(comparations)
    result.append(swaps)
    return result
```

```python
#n es el tamaño del arrelgo

def draw(n):
    listPermutations = []
    instructions = []
    comparations = []
    swaps = []


    for i in range(n):
        listPermutations.append(i)
    
    #Generando permutaciones
    permutations = (list(itertools.permutations(listPermutations)))

    print("Tamaño del arreglo: {}".format(n))
    print("Permutaciones: {}".format(len(permutations)))

    for i in permutations:
        accumulated = insertionSort(list(i))
        instructions.append(accumulated[0])
        comparations.append(accumulated[1])
        swaps.append(accumulated[2])

    #Graficas
    plt.hist(instructions, bins='auto')
    plt.title("Grafica de instruciones")
    plt.show()
    print("Promedio: {}".format(statistics.mean(instructions)))
    
    plt.hist(comparations, bins='auto')
    plt.title("Grafica de comparaciones")
    plt.show()
    print("Promedio: {}".format(statistics.mean(comparations)))

    plt.hist(swaps, bins='auto')
    plt.title("Grafica de intercambios")
    plt.show()
    print("Promedio: {}".format(statistics.mean(swaps)))
```

```python
draw(4)
```

```python
draw(8)
```

```python
draw(10)
```
