```python
import matplotlib.pyplot as plt
import timeit
```

```python
def fibonacci(n):
    
    if n <= 1:
        return 1
    
    fib = 1
    prevFib = 1
    
    for i in range(2, n):
        temp = fib
        fib += prevFib
        prevFib = temp
    
    return fib
```

```python
n = []
executionTimes = []
    
for i in range(1000):
    time = timeit.timeit("fibonacci({0})".format(i), setup="from __main__ import fibonacci",number = 1)
    n.append(i)
    executionTimes.append(time)
plt.plot(n, executionTimes)
plt.xlabel('n')
plt.ylabel('Tiempo de ejecucion')
plt.title('Tiempo de ejecucion del algoritmo de Fibonnaci')
plt.show()
```
