```python
import matplotlib.pyplot as plt
import statistics

listPermutations = []
```

```python
def swap(v, i, j):
    t = v[i]
    v[i] = v[j]
    v[j] = t
```

```python
def permutation(v, i, n):
    if i == n:
        v1 = v[:]
        listPermutations.append(v1)
    else:
        for j in range(i, n):
            swap(v, i, j)
            permutation(v, i + 1, n)
            swap(v, i, j)
```

```python
def insertionSort(array):

    comparations = 0
    swaps = 0
    instructions = 0
    for i in range(1,len(array)):
        temp = array[i]
        j = i - 1
        while j > -1 and temp < array[j]:
            comparations += 1
            array[j+1] = array[j]
            j = j-1
            swaps += 1
            instructions += 3
        instructions += 1
        array[j+1] = temp
        instructions += 4

        if j != i:
            comparations += 1

    instructions += 1
    result = []
    result.append(instructions)
    result.append(comparations)
    result.append(swaps)
    return result
```

```python
def draw(n):
    number = n
    array = list(i for i in range(number))
    instructions = []
    comparations = []
    swaps = []
    permutation(array, 0, number)

    for i in range(len(listPermutations)):
        accumulated = insertionSort(listPermutations[i])
        instructions.append(accumulated[0])
        comparations.append(accumulated[1])
        swaps.append(accumulated[2])

    
    plt.hist(instructions, bins='auto')
    plt.title("Histograma de instruciones")
    plt.show()
    print("Promedio: {}".format(statistics.mean(instructions)))

    plt.hist(comparations, bins='auto')
    plt.title("Histograma de comparaciones")
    plt.show()
    print("Promedio: {}".format(statistics.mean(comparations)))

    plt.hist(swaps, bins='auto')
    plt.title("Histograma de intercambios")
    plt.show()
    print("Promedio: {}".format(statistics.mean(swaps)))
```

```python
draw(4)
```

```python
draw(8)
```

```python
draw(10)
```
